```python
new_df = df[['id','name','host_id','host_name','neighbourhood_group','neighbourhood','room_type', 'price','minimum_nights',
            'number_of_reviews','calculated_host_listings_count','availability_365']]
new_df.head()
```

```python
hosts_areas = new_df.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].count().reset_index()
hosts_areas.sort_values(by='calculated_host_listings_count', ascending=False).head(5)
```

2. Price column is been used to check skewness in our dataset

```python
figsize=(12,8)
sns.boxenplot(x='price', data=new_df)
```

```python
areas_reviews = new_df.groupby(['name','neighbourhood_group','price','minimum_nights'])['number_of_reviews'].count().reset_index()
price_reviews1 = areas_reviews.sort_values(by='price', ascending=False).head(50)
price_reviews1
```

3. here we are visualizig price in compare to no. of reviews through simple bar plot

```python
area = price_reviews1['neighbourhood_group']
price = price_reviews1['price']

fig = plt.figure(figsize = (10,5))

#creating the bar plot
plt.bar(area, price, color ='maroon', width = 0.4)

plt.xlabel('area')
plt.ylabel('price')
plt.title('Price vs Number of reviews')
plt.show()
```

```python
price_reviews = areas_reviews.sort_values(by='number_of_reviews' , ascending=False).head(50)
price_reviews
```

```python
neighbourhood_group_count = new_df.groupby('neighbourhood_group').count().head()
neighbourhood_group_count
```

```python
new_df2 = new_df.groupby('neighbourhood')[['neighbourhood','host_name']].agg(['count'])['host_name'].sort_values(by='count',ascending=False).rename(index=str,columns={'Count':'Listing Count'})
```

```python
df2['RecruitmentSource'].value_counts().plot(kind='bar', color=['r','b','y','g'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(df2, x= df2['Sex'])
```

```python
plt.figure(figsize=(12,4))
sns.countplot(df2, x=df2['Department'], hue =df2['Sex'])
plt.title('Employee Gender diversity by Department')
```

```python
df2.columns
```

```python
df_relation = df2[['MarriedID', 'MaritalStatusID', 'GenderID',
       'EmpStatusID', 'DeptID', 'PerfScoreID', 'FromDiversityJobFairID',
       'Salary', 'Termd', 'PositionID','Sex','ManagerID', 'PerformanceScore', 'EngagementSurvey', 'EmpSatisfaction',
       'SpecialProjectsCount', 'DaysLateLast30','Absences' ]]
```

```python
plt.figure(figsize= (8,8))
sns.heatmap(df_relation.corr().round(2),annot = True, cmap ="YlGnBu")
```
